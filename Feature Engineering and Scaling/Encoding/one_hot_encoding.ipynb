# One hot encoding with top categories

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
df = pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
df.head()
```

```python
def replace_bool_with_int(value):
    if value == True:
        return 1
    elif value == False:
        return 0
    else:
        return value
```

```python
df = df.applymap(replace_bool_with_int)
df.head()
```
