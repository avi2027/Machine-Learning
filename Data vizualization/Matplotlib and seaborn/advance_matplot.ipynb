# 3d Line plot

```python
x = [0,5,10]
y = [0,15,18]
z = [0,23,30]

fig = plt.figure(figsize=(12,6))

ax = plt.subplot(projection='3d')

ax.scatter(x,y,z,marker = 'o',s = [100,100,100])
ax.plot3D(x,y,z,color = 'red')
```

# 3d surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,6))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap = 'winter')
plt.colorbar(p)
```

# contour plot

```python
# represet the 3d surface in 2d
fig = plt.figure(figsize=(12,6))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap = 'viridis')
plt.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,6))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap = 'viridis')
plt.colorbar(p)
```
