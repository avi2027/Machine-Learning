```python
import seaborn as sns
```

```python

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

# Categorical Plots
# Categorical Scatter Plot
* Stripplot
* Swarmplot
# Categorical Distribution Plots
* Boxplot
* Violinplot
# Categorical Estimate Plot -> for central tendency
* Barplot
* Pointplot
* Countplot

```python
# strip plot

sns.stripplot(x='day', y='total_bill', data=tips,jitter = False,cmap = 'viridis')
```

```python
sns.catplot(kind = 'strip',x='day', y='total_bill', data=tips,cmap = 'viridis',hue = 'sex')
```

```python
# swarm plot
sns.catplot(kind = 'swarm',x='day', y='total_bill', data=tips)
```

# box plot

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

# violinplot = boxplot + Kdeplot
* violinplot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

# Regression Plot

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,hue='sex')
```

```python
# resid plot
sns.residplot(x='total_bill', y='tip', data=tips)
```

# Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
iris
```

```python
sns.pairplot(iris,hue='species')
```

# pair grid

```python
g = sns.PairGrid(data = iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```
