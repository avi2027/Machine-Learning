```python
batsman = pd.read_csv("./datasets-session-23/sharma-kohli.csv")
batsman
plt.style.use('default')
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

# blue color virat kohli
# orange color rohit sharma
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red')

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 2)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',linestyle = 'dashed',linewidth = 2)

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")
```

```python
# using marker option
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 2,marker = 'D',markersize = 6)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',linestyle = 'dashed',linewidth = 2,marker = 'H',markersize = 6)

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")
```

```python
# individual labels
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 2,marker = 'D',markersize = 6,label = 'Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',linestyle = 'dashed',linewidth = 2,marker = 'H',markersize = 6,label = 'Rohit Sharma')

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")

plt.legend()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed',linewidth = 2,marker = 'D',markersize = 6,label = 'Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'red',linestyle = 'dashed',linewidth = 2,marker = 'H',markersize = 6,label = 'Rohit Sharma')

plt.title("Virat Kohli vs Rohit Sharma career comparison")
plt.xlabel("season")
plt.ylabel("runs scored")

plt.legend()
plt.grid()
```
